```python
#setting up a symptomps array:
filename = "2020_country_weekly_2020_GB_weekly_symptoms_dataset.csv";
data = pd.read_csv(filename).sort_values('date').groupby(['date']).mean()
```

```python
data.columns.map(lambda x: x.replace("symptom:", ""))[2:]
```

```python
data.plot(figsize=(25,35))
```

```python
#kw_list = ["dry eyes","headache", "diarea"]
kw_list = data.columns.map(lambda x: x.replace("symptom:", ""))[2:]
for kw in kw_list:
    pytrends.build_payload([kw], cat=0, timeframe='2019-11-30 2020-12-07', geo='', gprop='')
    if  overtime.empty:
        overtime = pytrends.interest_over_time()
    else:
        overtime[kw] = pytrends.interest_over_time()[kw].values
```
